Once we installed and imported all required packages we can start **scrapping** the specified Wikipedia page. 

```python
url = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
source = requests.get(url).text
    
soup = BeautifulSoup(source, 'lxml')

soup
```

As we can see the html file looks very messy and contains far more information then what we need for this assignment. Looking through the HTML file and having a look at the source code directly on the Wikipedia Page itself we can find out that the required information is contained in the *table* attribute within the HTML file.  

We'll use the **find_all** function from Beautifulsoup to extract only the elements contained within the *table* attribute.

```python
tables = soup.find_all('table')
tables
```

This is already far better!  

But still the format is still not quite the one we would like to get. We can see that the Headers we will use for the Table are specified as *th* whereas the rows we are interested in are preceded by *td*.  

Let's start by extracting the Headers from the HTML file and save them inside a list called **cols**. To do so, we will first search for all elements that start with *th* and then loop through these elements, eliminate the HTML code ('<th>' and '</th>), and then append these elements inside the list called **cols**.

```python
headers = tables[0].find_all('th')
cols = []

for header in headers: 
    cols.append(header.string.strip())
    
cols
```

The second step is now to do the same with the rows. We'll start by finding all elements starting with *td* then for each of these elements we strip them from the HTML artefacts and add them inside a list called **Canada_table**. 

```python
Canada_data = tables[0].find_all('td')
Canada_table = []

for data in Canada_data:
    Canada_table.append(data.string.strip())
```

However, we want to have a list where each element represents the full row. We know that the table will need to contain 3 attributes: Postal Code, Borough, Neighbourhood. Therefore, each row will contain exactly 3 elements.  

To slice the list into a list of lists containing each 3 elements we'll create the list called **canada_sublists** where each 3 elements from **Canada_table** will be read as a separate list and added to the **canada_sublists**. 

```python
canada_sublists = [Canada_table[i:i+3] for i in range(0, len(Canada_table), 3)]

canada_sublists
```

Once we have the two separate lists, **cols** containing the Column Headers of our DataFrame and the **canada_sublists** containing our rows, we can merge them into a DataFrame, that we will call **canada_df**.

```python
canada_df = pd.DataFrame(canada_sublists, columns=cols)

canada_df.head()
```

As mentioned in the assignment we will have to get rid of the lines in the DataFrame that have Boroughs that are *Not assigned*.  
To do so we will use the combination of the **str.contains** and the **~** to search for the rows that contain "Not assigned" and then substract them from the rest of the DataFrame.

```python
canada_df = canada_df[~canada_df['Borough'].str.contains("Not assigned")]

canada_df
```

Afterwards we still need to check if there are some Neighbourhoods that contain "Not assigned". 

```python
canada_df_unasgn = canada_df [canada_df['Neighbourhood'].str.contains("Not assigned")]

canada_df_unasgn
```

As you can see not Neighbourhood contained "Not assigned" as the DataFrame is empty. Therefore, we won't need to replace the name of unassigned Neighbourhoods.

To conclude let us print the shape of the DataFrame we could extract. 

```python
print(canada_df.shape)
```

**Foursquare API**  

Now we will use the Foursquare API to get more information about the venues in the identified Neighborhoods. Furthermore, we will use the information about the Category of venues provided by the Foursquare API.  

For this we will create a function called getNearbyVenues. 

```python
CLIENT_ID = 'NIDDKXAR0Y1JRGEEWFFQHC234U455BN21PQDFE3EFDRP1GJC' # your Foursquare ID
CLIENT_SECRET = '2FO3JXOSKZJLNJKOYE3L120JFVE1T2S5UKDCJC5NGP2PKVST' # your Foursquare Secret
VERSION = '20180605' # Foursquare API version
LIMIT = 100 # A default Foursquare API limit value

print('Your credentails:')
print('CLIENT_ID: ' + CLIENT_ID)
print('CLIENT_SECRET:' + CLIENT_SECRET)
```

```python
def getNearbyVenues(names, latitudes, longitudes, radius=500):
    
    venues_list=[]
    for name, lat, lng in zip(names, latitudes, longitudes):
        print(name)
            
        # create the API request URL
        url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
            CLIENT_ID, 
            CLIENT_SECRET, 
            VERSION, 
            lat, 
            lng, 
            radius, 
            LIMIT)
            
        # make the GET request
        results = requests.get(url).json()["response"]['groups'][0]['items']
        
        # return only relevant information for each nearby venue
        venues_list.append([(
            name, 
            lat, 
            lng, 
            v['venue']['name'], 
            v['venue']['location']['lat'], 
            v['venue']['location']['lng'],  
            v['venue']['categories'][0]['name']) for v in results])

    nearby_venues = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
    nearby_venues.columns = ['Neighborhood', 
                  'Neighborhood Latitude', 
                  'Neighborhood Longitude', 
                  'Venue', 
                  'Venue Latitude', 
                  'Venue Longitude', 
                  'Venue Category']
    
    return(nearby_venues)
```

Now that we have some information about the categories of venues present in the different Neighborhoods we can start understanding what kind of Neighborhoods we are dealing with.

For this we will try to map the frequency of these 240 unique venue categories and group them by Neighborhood.

```python
# one hot encoding
toronto_onehot = pd.get_dummies(toronto_venues[['Venue Category']], prefix="", prefix_sep="")

# add neighborhood column back to dataframe
toronto_onehot['Neighborhood'] = toronto_venues['Neighborhood'] 

# move neighborhood column to the first column
fixed_columns = [toronto_onehot.columns[-1]] + list(toronto_onehot.columns[:-1])
toronto_onehot = toronto_onehot[fixed_columns]

toronto_onehot.head(10)
```

```python
toronto_onehot.shape
```

```python
toronto_grouped = toronto_onehot.groupby('Neighborhood').mean().reset_index()

toronto_grouped
```

```python
toronto_grouped.shape
```

As we see we have a row by Neighborhood and a frequency value for each venue category.  

Now that we have this DataFrame we can group the top 5 venue categories by Neighborhood.

```python
num_top_venues = 5

for hood in toronto_grouped['Neighborhood']:
    print("----"+hood+"----")
    temp = toronto_grouped[toronto_grouped['Neighborhood'] == hood].T.reset_index()
    temp.columns = ['venue','freq']
    temp = temp.iloc[1:]
    temp['freq'] = temp['freq'].astype(float)
    temp = temp.round({'freq': 2})
    print(temp.sort_values('freq', ascending=False).reset_index(drop=True).head(num_top_venues))
    print('\n')
```

In order to get a DataFrame that list the 10 top most common venues by Neighborhood we'll start by creating a function called 'return_most_common_venues'.  

```python
def return_most_common_venues(row, num_top_venues):
    row_categories = row.iloc[1:]
    row_categories_sorted = row_categories.sort_values(ascending=False)
    
    return row_categories_sorted.index.values[0:num_top_venues]
```

```python
num_top_venues = 10

indicators = ['st', 'nd', 'rd']

# create columns according to number of top venues
columns = ['Neighborhood']
for ind in np.arange(num_top_venues):
    try:
        columns.append('{}{} Most Common Venue'.format(ind+1, indicators[ind]))
    except:
        columns.append('{}th Most Common Venue'.format(ind+1))

# create a new dataframe
neighborhoods_venues_sorted = pd.DataFrame(columns=columns)
neighborhoods_venues_sorted['Neighborhood'] = toronto_grouped['Neighborhood']

for ind in np.arange(toronto_grouped.shape[0]):
    neighborhoods_venues_sorted.iloc[ind, 1:] = return_most_common_venues(toronto_grouped.iloc[ind, :], num_top_venues)

neighborhoods_venues_sorted.head()
```

### 2.4 Cluster the Neighbourhoods

Now that we have a DataFrame that gives us some information about each Neighborhood analyzed we can start by running the k-means clustering to seperate the Neighborhoods by clusters.  

We will set the number of clusters at 5.  

```python
# set number of clusters
kclusters = 5

toronto_grouped_clustering = toronto_grouped.drop('Neighborhood', 1)

# run k-means clustering
kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(toronto_grouped_clustering)

# check cluster labels generated for each row in the dataframe
kmeans.labels_[0:10] 
```

As we now have an array of clusters let's add this information to the neighborhood_venues_sorted DataFrame. Once we did this we can join this DataFram with the general *toronto_data*.  

```python
# add clustering labels
neighborhoods_venues_sorted.insert(0, 'Cluster Labels', kmeans.labels_)

toronto_merged = toronto_data

# merge toronto_grouped with toronto_data to add latitude/longitude for each neighborhood
toronto_merged = toronto_merged.join(neighborhoods_venues_sorted.set_index('Neighborhood'), on='Neighborhood')

toronto_merged.head()
```

Now let's visualize the Clusters on the map. 

```python
# create map
map_clusters = folium.Map(location=[latitude, longitude], zoom_start=11)

# set color scheme for the clusters
x = np.arange(kclusters)
ys = [i + x + (i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, poi, cluster in zip(toronto_merged['Latitude'], toronto_merged['Longitude'], toronto_merged['Neighborhood'], toronto_merged['Cluster Labels']):
    label = folium.Popup(str(poi) + ' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
